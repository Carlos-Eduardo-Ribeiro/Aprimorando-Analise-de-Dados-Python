```python
alunos = []

while True:
    print("\n=== Cadastro de Alunos ===")
    nome = input("Nome do aluno: ")
    idade = int(input("Idade: "))

    notas = []
    for i in range(1, 4):
        nota = float(input(f"Nota {i}: "))
        notas.append(nota)

    media = sum(notas) / 3

    aluno = {
        "nome": nome,
        "idade": idade,
        "notas": notas,
        "media": media
    }

    alunos.append(aluno)

    opc = input("Deseja cadastrar outro aluno? (s/n): ").lower()
    if opc != "s":
        break

# ——— Relatórios finais ———
print("\n=== Lista de Alunos Cadastrados ===")
for a in alunos:
    print(f"{a['nome']} | Idade: {a['idade']} | Notas: {a['notas']} | Média: {a['media']:.2f}")

# Maior média
melhor = max(alunos, key=lambda x: x["media"])
# Menor média
pior = min(alunos, key=lambda x: x["media"])

print("\nAluno com MAIOR média:")
print(f"{melhor['nome']} — Média: {melhor['media']:.2f}")

print("\nAluno com MENOR média:")
print(f"{pior['nome']} — Média: {pior['media']:.2f}")
```
